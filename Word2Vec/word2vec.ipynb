```python
import os
import nltk
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
from gensim.models import Word2Vec
```

```python
file_path = '/home/aadityasukhoi/Desktop/GLA/NLP/Word2Vec/archive/alllines.txt'
assert os.path.exists(file_path), f"Not Found: {file_path}"

with open(file_path, 'r', encoding='utf-8') as f:
    corpus = f.read()
    
print("Preview:\n", corpus[:500])
print("Corpus Length (chars):", len(corpus))
```

```python
nltk.download("punkt", quiet=True)
raw_sentences = sent_tokenize(corpus)
sentences = [simple_preprocess(s) for s in raw_sentences]
print("Number of sentences: ", len(sentences))
print("Example tokenized sentence: ", sentences[0][:20])
```

```python
len(corpus)
```

```python
model = Word2Vec(
    vector_size=100, # dimension of word vectors
    window=10, # context window size
    min_count=2, # ignore words with frequency < 2
    workers=4 
)
```

```python
model.build_vocab(sentences)
```

```python
model.train(sentences, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar("children")
```

```python
model.wv.doesnt_match(["king", "queen", "breathe"])
```
