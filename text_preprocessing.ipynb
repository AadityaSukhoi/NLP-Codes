<b>i. HTML Tag Removal</b>

```python
text1 = "<p>The <b>movie</b> was <i>fantastic</i> and <a href='#'>enjoyable</a>.</p>"
```

```python
import re #Using Regex
def remove_html_tags(text):
    pattern = re.compile(r'<.*?>')
    return pattern.sub(r'', text)
```

```python
remove_html_tags(text1)
```

```python
import re
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\S.\S+')
    return pattern.sub(r'', text)
```

<b>iii. Remove Punctuation</b>

```python
import string
string.punctuation
```

```python
exclude = string.punctuation
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char, '')
    return text
```

```python
text5 = "remove. with, punctuations!?"
```

```python
print(remove_punc(text5))
```

```python
def remove_stopwords(text):
    new_text = []

    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    
    x=new_text[:]
    new_text.clear()
    return " ".join(x)
```

<b>v. Stemming / Lemmetization</b>

```python
# Stemming: extracts the root of the word but does not ensure that the root word is a part of the english language or the dictionary
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walking walked walks walk"
```

```python
stem_words(sample)
```

```python
# Lemmetization: extracts the root of the word and ensures that the root word is a part of the english language or the dictionary
import nltk 
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

sentence = "He was running and eating at the same time, but now he has a bad habit of swimming late at night and ignoring his friends' advice."
punctuations = "?:!.,;"
```

```python
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)

sentence_words
```

```python
print("{0:20} {1:20}".format("Word", "Lemma"))
for word in sentence_words:
    print("{0:20} {1:20}".format(word, wordnet_lemmatizer.lemmatize(word, pos='v'))) 
```

```python
# subword tokenization

import sentencepiece as spm

# Create a tiny corpus file
with open("corpus.txt", "w", encoding="utf-8") as f:
    f.write("""
    Natural Language Processing is interesting.
    Subword tokenization helps handle unknown words.
    Unhappiness can be split into meaningful parts.
    Transformers use subword tokenizers.
    """)
```

```python
spm.SentencePieceTrainer.train(
    input="corpus.txt",
    model_prefix="subword_model",
    vocab_size=100,
    model_type="bpe"   # Byte Pair Encoding
)
```

```python
sp = spm.SentencePieceProcessor()
sp.load("subword_model.model")
```

```python
text = "unhappiness"

subword_tokens = sp.encode_as_pieces(text)
print("Subword Tokens:", subword_tokens)
```

```python
sentence = "Subword tokenization is powerful"
print(sp.encode_as_pieces(sentence))
```

<b>vi. Language Detection</b>

!pip install langdetect

```python
text_clean2 = "La pelicula fue absolutamente fantástica y agradable!"
```

```python
from langdetect import detect, detect_langs

def detect_language(text):
    try:
        lang = detect(text)
        lang_prob = detect_langs(text)
        return lang, lang_prob
    except Exception as e:
        return "Error", str(e)
```

```python
# Example usage
text_clean2 = "La pelicula fue absolutamente fantástica y agradable!"
lang_code, lang_prob = detect_language(text_clean2)

print("Text: ", text_clean2)
print("Detected Language Code: ", lang_code)
print("Confidence: ", lang_prob) # es-Spanish
```
