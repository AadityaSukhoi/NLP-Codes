```python
import re

# match  a word at the beginning of a string
result = re.match("The", r'The cat jumped gracefully onto the windowsill, chasing a beam of sunlight.')
print(result)
```

```python
print(result.group())  # Output: The
```

```python
result =  re.search("bright", "The sun is bright. She wore bright colors.")
print(result.group())  # Output: bright
```

```python
result = re.findall('bright', "The sun is bright. She wore bright colors.")
print(result)  # Output: ['bright', 'bright']
```

```python
str = r"The cat jumped gracefully onto the windowsill, chasing a beam of sunlight."
```

```python
x = re.findall(r"\bjum", str)
print(x)  # Output: ['jum']
```

```python
str = "Today's date is 15 January 2025."
```

```python
x = re.findall(r"\d", str)
print(x)  

if(x):
    print("There is a match.")
else:
    print("No match found.")
```

```python
x = re.findall(r"\d+", str)
print(x)  

if(x):
    print("There is a match.")
else:
    print("No match found.")
```

```python
x = re.findall(r"\D", str)
print(x)  

if(x):
    print("There is a match.")
else:
    print("No match found.")
```

```python
x = re.findall(r"\D+", str)
print(x)  

if(x):
    print("There is a match.")
else:
    print("No match found.")
```
